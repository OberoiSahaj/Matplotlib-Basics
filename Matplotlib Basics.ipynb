```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line Plot

```python
# let's suppose this is our data.
plt.style.use('seaborn')
x = np.arange(1,6) 
y1 = x**2
y2 = x**3 #np.arange(10,31,5)
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x, y1, color= 'blue', label= 'Squares', marker = 'o')
plt.plot(x, y2, color = 'red', label= 'Cubes', linestyle='dashed', marker= '*')
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Y Numbers')
plt.title('Line Plot')
plt.show()
```

```python
plt.plot(y2) # indices of elements of y2 as x
```

# Scatter Plot

```python
plt.scatter(x, y1, color= 'blue', label= 'Squares', marker = '*')
plt.scatter(x, y2, color = 'red', label= 'Cubes',  marker= '^')
plt.legend()
plt.xlabel('Numbers')
plt.ylabel('Y Numbers')
plt.title('Line Plot')
plt.show()
```

# Bar Plots

```python
x = np.arange(1,6) 
plt.bar(x, y2, width=.5) #by default width = 0.8
plt.show()
```

```python
x = np.arange(1,6)
plt.bar(x, y2) #by default width = 0.8
plt.bar(x, y1)
plt.show()
```

```python
x = np.arange(1,6) * 2 
plt.bar(x, y1, width=.5, label='Last yr', tick_label = ['potato', 'tomato','apple','pear', 'capsicum']) #by default width = 0.8
plt.bar(x + 0.5, y2, width=.5, label='Current yr')
plt.title('Vegetable, Fruit price hike')
plt.ylabel('Prices')
plt.xlim(0,12)
plt.ylim(0,130)
plt.legend()
plt.show()
```

# Pie Charts

```python
subjects = ['Player sales', 'Ticket prices', 'Shirt sales', 'Sponsors']
marks = [30, 10, 5, 20]
```
